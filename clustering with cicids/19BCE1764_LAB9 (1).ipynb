```python
from IPython.display import Image
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
%matplotlib inline

%matplotlib inline
```

```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/datasets/train_MachineLearningCVE.csv');
X = data.iloc[:,1:3]
X = X.values
X
```

```python
plt.scatter(X[:,0],X[:,1], c='red', marker = 'o', s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='random', n_init=1, max_iter=2, tol=1e-04, random_state=0).fit(X)
kmeans.fit(X)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python
plt.scatter(X[labels == 0,0], X[labels == 0, 1], s= 50, c= 'lightgreen', marker='s', label='cluster 1')
plt.scatter(X[labels == 1,0], X[labels == 1, 1], s= 50, c= 'orange', marker='o', label='cluster 2')
plt.scatter(X[labels == 2,0], X[labels == 2, 1], s= 50, c= 'lightblue', marker='v', label='cluster 3')

plt.scatter(centroids[:, 0], centroids[:, 1], s = 250, marker='*', c = 'red', label='centroids')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#Elbow Method
distortions = []
for i in range(1, 11):
  km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
  km.fit(X)
  distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortions')
plt.tight_layout()
plt.show()
```
